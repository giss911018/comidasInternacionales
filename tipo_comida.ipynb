```python
import datetime
import random

def get_meal_suggestions():
    current_hour = datetime.datetime.now().hour

    meal_times = {
        "desayunar": {
            "start": 6, "end": 10,
            "primer_plato": [
                "Huevos revueltos con tostadas y aguacate (México/Latinoamérica)",
                "Avena con frutos secos y miel (Reino Unido/Internacional)",
                "Tostadas de aguacate con tomate y semillas (Vegano)"
            ],
            "segundo_plato": [
                "Panqueques con jarabe de arce y frutas del bosque (Estados Unidos/Canadá)",
                "Gallo Pinto con Natilla (Costa Rica)",
                "Croissant con mantequilla y mermelada (Francia)"
            ],
            "postre": [
                "Fruta fresca de temporada",
                "Yogur griego con miel y nueces"
            ],
            "bebida": [
                "Café con leche",
                "Té verde",
                "Zumo de naranja natural"
            ]
        },
        "almorzar": {
            "start": 12, "end": 15,
            "primer_plato": [
                "Ensalada César con pollo a la parrilla (México/Estados Unidos)",
                "Sopa de lentejas con pan integral (España/Medio Oriente)",
                "Lentejas con verduras y arroz integral (Vegano)"
            ],
            "segundo_plato": [
                "Pasta con salsa pesto y tomates cherry (Italia)",
                "Curry de pollo con arroz (India)",
                "Tacos al pastor (México)"
            ],
            "postre": [
                "Fruta cortada",
                "Flan de vainilla",
                "Mousse de chocolate"
            ],
            "bebida": [
                "Agua fresca de Jamaica",
                "Limonada natural",
                "Vino tinto (copa)"
            ]
        },
        "merendar": {
            "start": 16, "end": 18,
            "primer_plato": [
                "Yogur con granola y frutas (Internacional)",
                "Hummus con bastones de zanahoria y pepino (Vegano)"
            ],
            "segundo_plato": [
                "Sándwich pequeño de jamón y queso (Internacional)",
                "Empanadas (Argentina/Latinoamérica)"
            ],
            "postre": [
                "Té con galletas (Reino Unido)",
                "Muffin de arándanos"
            ],
            "bebida": [
                "Batido de frutas con espinacas (Internacional)",
                "Café con leche",
                "Infusión de manzanilla"
            ]
        },
        "cenar": {
            "start": 19, "end": 22,
            "primer_plato": [
                "Crema de calabaza (Internacional)",
                "Ensalada mixta (Internacional)"
            ],
            "segundo_plato": [
                "Salmón al horno con espárragos (Nórdico/Internacional)",
                "Pechuga de pollo a la plancha con puré de patatas (Internacional)",
                "Tortilla española con ensalada (España)",
                "Sushi variado (Japón)",
                "Pizza margherita (Italia)",
                "Curry de garbanzos y espinacas con arroz basmati (Vegano)"
            ],
            "postre": [
                "Helado de vainilla",
                "Fruta de temporada",
                "Tarta de queso"
            ],
            "bebida": [
                "Agua mineral",
                "Cerveza artesanal",
                "Vino blanco (copa)"
            ]
        }
    }

    current_meal_time = None
    for meal, data in meal_times.items():
        if data["start"] <= current_hour < data["end"]:
            current_meal_time = meal
            break

    if current_meal_time:
        print(f"¡Es hora de {current_meal_time}! Vamos a construir tu menú personalizado:")
        personalized_menu = {}
        meal_data = meal_times[current_meal_time]

        categories_map = {
            "primer_plato": "Primer Plato",
            "segundo_plato": "Segundo Plato",
            "postre": "Postre",
            "bebida": "Bebida"
        }

        for category_key, category_name in categories_map.items():
            if category_key in meal_data and meal_data[category_key]:
                options = list(meal_data[category_key]) # Create a mutable copy
                current_selection = None
                while current_selection is None:
                    suggestion = random.choice(options)
                    response = input(f"  ¿Te gustaría '{suggestion}' como {category_name}? (s/n/salir para terminar): ").lower()
                    if response == 's':
                        current_selection = suggestion
                        personalized_menu[category_name] = suggestion
                    elif response == 'n':
                        if len(options) > 1:
                            options.remove(suggestion) # Remove used suggestion to avoid repetition
                        else:
                            print(f"    No hay más opciones disponibles para {category_name}. Seleccionando: {suggestion}")
                            current_selection = suggestion
                            personalized_menu[category_name] = suggestion
                    elif response == 'salir':
                        print(f"    Saltando la selección para {category_name}.")
                        personalized_menu[category_name] = "No seleccionado"
                        current_selection = "skipped" # Mark as skipped to exit inner loop
                    else:
                        print("    Respuesta no válida. Por favor, responde 's' para sí, 'n' para no, o 'salir'.")
            else:
                personalized_menu[category_name] = "No hay opciones disponibles"
                print(f"  No hay opciones disponibles para {category_name}.")

        print("\n--- Tu Menú Personalizado --- ")
        for course, item in personalized_menu.items():
            print(f"{course}: {item}")

    else:
        print("Ahora mismo no es una hora de comida principal definida. ¡Disfruta tu día!")

# Llamar a la función para obtener las sugerencias de comidas
get_meal_suggestions()
```
